```python
#딕셔너리는 어떻게 만들까?
dic = {"name" : "ken", "phone" : "010-1234-1234", "birth" : "00-01-01"}
```

```python
#딕셔너리 쌍 추가, 삭제하기
dic["hobby"] = "running"
dic
```

```python
dic[3] = [1, 2, 3]
dic
```

```python
#딕셔너리 요소 삭제하기
del dic[3]
dic
```

```python
del dic["hobby"]
dic
```

```python
# 딕셔너리 요소 변경하기
dic["name"] = "python"
dic
```

```python
#딕셔너리에서 key를 사용해서 value 얻기
print(dic["name"])
```

```python
#딕셔너리 관련 함수
##key 리스트 만들기 - keys
dic = {
    "name" : "ken", 
    "phone" : "010-1234-1234", 
    "birth" : "00-01-01"
}


list(dic.keys())
```

```python
#VAlue 리스트 만들기 - values
li = dic.values()
list(li)
```

```python
#key, value 쌍으로 얻기 - items
li = dic.items()
list(li)
```

```python
#key : value 쌍으로 모두 지우기 - clear
dic.clear()
dic
```

```python
#key로 alue 얻기 - get
dic = {
    "name" : "ken", 
    "phone" : "010-1234-1234", 
    "birth" : "00-01-01"
}
# dic["name"]
print(dic.get('name'))
print(dic.get('phone'))
print(dic.get('birth'))
```

```python
#해당 key가 딕셔너리 안에 있는지 조사하기 - in
dic = {
    "name" : "ken", 
    "phone" : "010-1234-1234", 
    "birth" : "00-01-01"
}

print("name" in dic)
print("email" in dic)
```

```python
#key로 value 얻기 - pop
dic = {
    "name" : "ken", 
    "phone" : "010-1234-1234", 
    "birth" : "00-01-01"
}
phone = dic.pop('phone')
print(phone)
```
