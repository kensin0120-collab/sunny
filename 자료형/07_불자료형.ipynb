```python
#자료형의 참과 거짓
print(bool("python"))
print(bool(""))

print(bool([1, 2, 3]))
print(bool([]))

print(bool((1, 2, 3)))
print(bool(()))

print(bool({"a" : 1}))
print(bool({}))


print(bool(1))
print(bool(0))


print(bool(None))

```

```python
#and 연산자, or 연산자

print(True and True)
print(True and False)
print(False and False)
print(False and True)

print(True or True)
print(True or False)
print(False or False)
print(False or True)


```
