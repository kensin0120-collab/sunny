```python
#돈이 3000원 이상 있거나 카드가 있다면 택시를 타고 가고, 그렇지 않으면 걸어가라.

money = 2000
card = True

if money >=3000 or card == True:
    print("택시를 타고 가라")
else:
    print("걸어가라")
```

```python
#만약에 주머니에 money가 있으면 택시를 타고 가고, 그렇지 않으면 걸어가라.

pocket = ["paper", "phone", "money"]

if money in pocket:
    print("택시를 타고 가라")
else:
    print("걸어가라")
```

```python
# 다양한 조건을 판단하는 elif
pocket = ["paper", "phone"]
card = True

if "money" in pocket:
    print("택시를 타고 가라")
else:
    if card == True:
        print("택시를 타고 가라")
    else:
        print("걸어가라")
```

```python
# 조건부 표현식
score = 85

if score >= 60:
    result = "합격"
else:
    result = "불합격"

print(result)
```

```python
# 학점이 90이상이면 A, 
# 80이상 90미만이면 B,
# 70이상 80미만이면 C,
# 60이상 70미만이면 D,
# 60미만이면 F,
# 출력하는 프로그램을 만드시오.


score = int(input("학점을 입력하세요 : "))

if score >= 90 :
   result = (f"입력값은 {score}, 당신의 학점은: A")
elif score >= 80 :
   result = (f"입력값은 {score}, 당신의 학점은: B")
elif score >= 70 :
   result = (f"입력값은 {score}, 당신의 학점은: C")
elif score >= 60 :
   result = (f"입력값은 {score}, 당신의 학점은: D")
elif score < 60 :
   result = (f"입력값은 {score}, 당신의 학점은: F")

print(result)
```
