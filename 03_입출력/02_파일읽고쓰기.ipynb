```python
f = open("data2.txt", "w")
f.close()

# 파일객체 = open(파일이름, 파일열기모드)

# r - 읽기모드 : 파일을 읽기만 할 때
# w - 쓰기모드 : 파일의 내용을 쓸 때
# a - 추가모드 : 파일의 마지막에 내용을 추가 할 때
```

```python
f = open("./data/새파일.txt", "w")

for i in range(1, 11):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)

f.close()
```

```python
f = open("data/새파일.txt", "r")
while True:
    line = f.readline()
    if not line : break
    print(line)
f.close()
```

```python
# readlines() 함수 이용하기
f = open("data/새파일.txt", "r")
lines = f.readlines()

for line in lines:
    print(line, end="")
f.close()

# print(type(lines))
# print(len(lines))
```

```python
# read 함수 사용하기
f = open("data/새파일.txt", "r")
data = f.read()
print(data)
f.close()
```

```python
# 파일에 새로운 내용 추가하기
f = open("../새파일.txt", "a")

for i in range(11, 21):
    f.write(f"{i}번째 줄입니다.\n")

f.close()
```

```python
# with문과 함께 사용하기(클로즈를 안 해도 됨)
with open("data2.txt", "w") as f:
    f.write("Life is too short, You need Python.")
```
